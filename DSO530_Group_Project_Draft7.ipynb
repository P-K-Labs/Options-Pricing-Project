```python
# create vars S/K and exp(tau*r) : referred to Black-Scholes model

df["price_ratio"] = df["S"]/df["K"]

df['S_r'] = df['S']*df['r']
df['K_r'] = df['K']*df['r']
#df['Value_r'] = df['Value'] * df['r']
df['S_t'] = df['S']*df['tau']
df['K_t'] = df['K'] * df['tau']
#df['Value_t'] = df['Value']*df['tau']
df['r_by_t'] = df['r']/df['tau']
#df['Value/k'] = df['Value']/df['K']
df["price_ratio"] = df["S"]/df["K"]
df["r_t"] = df["tau"]*df["r"]
df['asset_val'] =( df['S_r'] + df['S'])**df['tau']
df['asset_val_delta'] =( df['K'] - df['asset_val'])/df['asset_val']
```
